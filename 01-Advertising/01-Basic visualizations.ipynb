```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 

%matplotlib inline
```

```python
# change this to match the path to your dataset
path_to_data = '../Data/Advertising.csv'
```

```python
df = pd.read_csv(path_to_data)
df = df.drop(columns=['Unnamed: 0'])
df.head(3)
```

```python
plt.figure(figsize=(15, 7))
plt.subplot(1, 3, 1)
plt.scatter(df['TV'], df['sales'])
plt.xlabel('TV')
plt.ylabel('sales')
plt.subplot(1, 3, 2)
plt.scatter(df['radio'], df['sales'])
plt.xlabel('radio')
plt.ylabel('sales')
plt.subplot(1, 3, 3)
plt.scatter(df['newspaper'], df['sales'])
plt.xlabel('newspaper')
plt.ylabel('sales')
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.subplot(1, 3, 1)
sns.regplot(x=df['TV'], y=df['sales'], color="r")
plt.subplot(1, 3, 2)
sns.regplot(x=df['radio'], y=df['sales'], color="g")
plt.subplot(1, 3, 3)
sns.regplot(x=df['newspaper'], y=df['sales'], color="y")
plt.show()
```
